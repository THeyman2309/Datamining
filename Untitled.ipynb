##### normalized stress


```python
# sum((high_dim_i - low_dim_i)**2)/sum(high_dim_i) for i in data
from sklearn.metrics import pairwise

def normalized_stress(data_high, data_low):
    # get pairwise distances high_dimensional data
    high_dist = pairwise.euclidean_distances(data_high)

    # get pairwise distances low_dimensional data
    low_dist = pairwise.euclidean_distances(data_low)
    
    l = data_high.shape[0]
    total = 0
    # get sum of pairwise distances data_high
    for i in range(l):
        for j in range(i + 1,):
            total += high_dist[i][j] ** 2
        
    stress = 0
    # get stress
    for i in range(l):
        for j in range(i + 1,):
            stress += (high_dist[i][j] - low_dist[i][j]) ** 2
    # normalise stress
    return stress / total
    
    
```

## Clustering

#### Does the data have clustering tendencies?

If the hopkins score is above 0.75 the data shows clustering tendencies

```python
# taken from https://sushildeore99.medium.com/really-what-is-hopkins-statistic-bad1265df4b
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan
def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n)
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)

    rand_X = sample(range(0, n, 1), m)

    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0

    return H
```

```python
# get labels for dataset3
# read in labels
meta_data = pd.read_csv("C:\\Users\\tobia\\Documents\\E-MTAB-11134\\E-MTAB-11134.sdrf.txt", low_memory=0,delimiter='\t')

# make dictionary
dict_labels = {meta_data.iloc[i,0]: meta_data.iloc[i,5] for i in range(57)}
labels = [dict_labels[i] for i in list(dataset3.index)]
```

```python
# make functions to evaluate clustering
from sklearn.metrics.pairwise import euclidean_distances

def compute_incidence(clustering_array):
    x = clustering_array
    X = np.tile(x, (x.shape[0], 1))
    Xt = X.transpose()
    incidence = X == Xt
    return(incidence)


def correlation(reduction, cluster):
    
    # make proximity matrix and incidence matrix
    proximity_matrix = euclidean_distances(reduction, reduction)
    incidence_matrix = compute_incidence(cluster)

    # convert matrices to arrays
    proximity_matrix_flat = proximity_matrix.flatten()
    incidence_matrix_flat = incidence_matrix.flatten()

    # get correlation (cor ~= 0.5 => no clustering tendencies)
    pcc = np.corrcoef(proximity_matrix_flat, incidence_matrix_flat)
    return pcc[0,1]



# accuracy based on label (TP + TN)/(TN + FN + FP + TP)
def accuracy(cluster, labels):
    TP, FP, TN, FN = 0, 0, 0, 0

    for i in range(57):
        if cluster[i] == 1 and labels[i] == "inflammatory breast carcinoma" :
            TP += 1
        elif cluster[i] == 1 and labels[i] == "non-inflammatory breast carcinoma":
            FP += 1
        elif labels[i] == "inflammatory breast carcinoma" :
            FN += 1
        elif labels[i] == "non-inflammatory breast carcinoma":
            TN += 1
    return ((TP + TN)/(TN + FN + FP + TP), TP, FP, TN, FN)
compute_incidence(np.array([1,1,1,1,0,0,0]))
```

```python
# for random data
from sklearn.cluster import KMeans
import random
corr = []
acc = []
for i in range(100):
    # create random points
    data = pd.DataFrame([[random.random(), random.random()] for i in range (100)])
    # add labels randomly
    rand_lab = [i for i in labels]
    random.shuffle(rand_lab)
    # create random clustering
    clusters = KMeans(n_clusters=2).fit_predict(data)
    ac = accuracy(clusters, rand_lab)[0]
    cor = correlation(data, clusters)
    corr.append(cor)
    acc.append(ac)
    
    # adding values to rest of the data
    value.append(ac)
    dr.append("random")
    metric.append("accuracy")
    value.append(cor)
    dr.append("random")
    metric.append("correlation")
    
print(np.mean(corr), np.std(corr))
print(np.mean(acc), np.std(acc))
    
```
